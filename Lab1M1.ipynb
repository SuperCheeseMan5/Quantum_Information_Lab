```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import numpy as np

# Simulate 1000 qubits measurements (|0> = 60%, |1> = 40%)
outcomes = np.random.choice([0,1], size=1000, p=[.6,.4])
plt.figure(figsize=(8,5))
plt.hist(outcomes, bins=2, rwidth=.8, alpha=.7, color=['skyblue'])
plt.xlabel("Measurement Outcome (|0> vs |1>)")
plt.ylabel("Counts")
plt.title("Simulated Single Qubit Measurements")
plt.xticks([0.25,0.75], ['|0>', '|1>'])
plt.grid(axis='y', alpha=0.3)
plt.show()
```

```python
def get_n_qubit_measurements(n_qubits=2, n_runs=1024, prob=[.5]*n_qubits):
    """ Take a number of measurements for an n-qubit system with a given probability

    Keyword arguments:
        n=qubit
    """
    # Initialize an array to store the results for each qubit
    qubits = [0]*n_qubits
    # Using a for loop get the measurement for each qubit
    for q in range(0, n_qubits):
        qubits[q] = np.random.choice(['0','1'], size = n_runs, p = [1-prob[q],prob[q]])
    # reformat the array to group each qubit run together
    measurements = []
    for i in range(0, n_runs):
        run = ""
        for q in range(0, n_qubits):
            run = run +qubits[q][i]
        measurements.append(np.array(run))
    return np.array(measurements)
```

```python
n_qubits = 3
n_runs = 1024
probabilities = [.7,.2,.1]
# probabilities = [.5,.5,.5]
```

```python
measurements = get_n_qubit_measurements(n_qubits, n_runs, probabilities)
unique, counts = np.unique(measurements, return_counts=True)
plt.figure(figsize=(10,6))
plt.bar(unique, counts, alpha=0.7, color=['maroon'])
plt.xlabel(f'{n_qubits}-Qubit State Outcome')
plt.ylabel('Counts')
plt.title(f'Simulate {n_qubits}-Qubit Measurement ({n_runs} runs)')
plt.show()
```
