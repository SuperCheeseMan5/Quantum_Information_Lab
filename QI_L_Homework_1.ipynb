<h4> Libraries </h4>

```python
# Import Libraries
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import numpy as np
```

<h4> Methods </h4>

```python
# Define local variables (Physics Stuff!)
n_qubits = 3
n_runs = 1024
probabilities = [.7,.2,.1]
```

```python
# Randomly generate 3 different 3 qubit states with the following probabilities
# |000> -> 70%, |101> -> 20%, |111> -> 10% 
measurements = np.random.choice(['000','101', '111'], size = n_runs, p = probabilities)
# Seperate the list of 'measurements' by the qubit sting and how many of each there is
unique, counts = np.unique(measurements, return_counts=True)
```

<h4> Results </h4>

```python
# Creating the bar graph from measurements.
plt.figure(figsize=(10,6))
plt.bar(unique, counts, alpha=0.7, color=['maroon'])
plt.xlabel(f'{n_qubits}-Qubit State Outcome')
plt.ylabel('Counts')
plt.title(f'Simulate {n_qubits}-Qubit Measurement ({n_runs} runs)')
plt.show()
```

```python
# Actual probabilities of results
print(f"Measurement | Expected Probability | Result Probability |")
for i in range(0, len(counts)):
    print(f"   |{unique[i]}>    |         {probabilities[i]:2.0%}          |        {counts[i]/n_runs:.2%}      |")
```

<p>The results of the measurements produced by <code>np.random</code> result in close probabilites to the ones given but not exact. This accuracy can be increased or degreased by changing the ammount or runs or counts preforemed. A higher number of runs should increase the probabilities accuracy.</p> 

```python
# Remake the chart with more runs
n_runs2 = n_runs * 500 
measurements2 = np.random.choice(['000','101', '111'], size = n_runs2, p = probabilities)
unique2, counts2 = np.unique(measurements2, return_counts=True)
print(f"Measurement | Expected Probability | Result Probability |")
for i in range(0, len(counts2)):
    print(f"   |{unique2[i]}>    |         {probabilities[i]:2.0%}          |        {counts2[i]/n_runs2:.2%}      |")
```
